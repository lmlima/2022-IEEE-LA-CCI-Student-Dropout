# Dados originais

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/df_original.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.size
```

```python
df = df.fillna(0)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

x = df.loc[:, df.columns].values
x = StandardScaler().fit_transform(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_df = pd.DataFrame(x,columns=feat_cols)
normalised_df.tail()
```

```python
from sklearn.manifold import TSNE

tsne_evasao = TSNE(n_components=2, n_iter=1000, perplexity=100, verbose=2).fit_transform(x)
```

```python
tsne_evasao_df = pd.DataFrame(data = tsne_evasao,
                              columns = ['x', 'y'])

tsne_evasao_df.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.title("t-SNE",fontsize=20)
targets = [0, 1]
colors = ['green', 'blue']
for target, color in zip(targets, colors):
    indicesToKeep = df['LABEL'] == target
    plt.scatter(tsne_evasao_df.loc[indicesToKeep, 'x'],
                tsne_evasao_df.loc[indicesToKeep, 'y'], edgecolors=color, s = 10, alpha=0.5, facecolors='none')

plt.legend(targets,prop={'size': 15})
```
